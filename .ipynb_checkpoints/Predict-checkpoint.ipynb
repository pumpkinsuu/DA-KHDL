# Predict

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import progressbar

from sklearn.metrics import explained_variance_score, mean_absolute_error, median_absolute_error
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression, LinearRegression

import matplotlib.pyplot as plt
```

```python
def calculate_accuracy(pred, y):
    tp = np.sum(np.logical_and(pred, y))
    fp = np.sum(np.logical_and(pred, np.logical_not(y)))
    tn = np.sum(np.logical_and(np.logical_not(pred), np.logical_not(y)))
    fn = np.sum(np.logical_and(np.logical_not(pred), y))

    tpr = 0 if (tp + fn == 0) else float(tp) / float(tp + fn)
    fpr = 0 if (fp + tn == 0) else float(fp) / float(fp + tn)
    acc = float(tp + tn) / len(pred)
    
    print(f'Accuracy: {acc: .2f}')
    print(f'TPR: {tpr: .2f}')
    print(f'FPR: {fpr: .2f}')
```

```python
#df = pd.read_csv('c_data/data.csv')
df = pd.read_csv('c_data_html/data.csv')
df
```

**Dự đoán nhiệt độ trung bình:**
* Chia dataset 80% train và 20% test
* Sử dụng Linear Regression

```python
columns = ['y_temp', 'y_temp_min', 'y_temp_max', 'y_rain', 'y_is_rain']
X = df[[col for col in df.columns if col not in columns]]
y = df[columns[0]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

print("The Explained Variance: %.2f" % regressor.score(X_test, y_test))
print("The Mean Absolute Error: %.2f degrees celsius" % mean_absolute_error(y_test, y_pred))
print("The Median Absolute Error: %.2f degrees celsius" % median_absolute_error(y_test, y_pred))
```

**Tính độ chính xác:**

```python
calculate_accuracy(y_pred, np.array(y_test))
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(211)
ax.set_ylabel('Temp')
ax.set_title('Predict vs Actual')

ax.plot(np.array(y_test))
ax.plot(y_pred)
```

**Dự đoán trời mưa:**
* Chia dataset 80% train và 20% test
* Sử dụng Logistic Regression

```python
columns = ['y_temp', 'y_temp_min', 'y_temp_max', 'y_rain', 'y_is_rain']
X = df[[col for col in df.columns if col not in columns]]
y = df[columns[4]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf = LogisticRegression(max_iter=10000)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

**Tính độ chính xác:**

```python
calculate_accuracy(y_pred, np.array(y_test))
```
